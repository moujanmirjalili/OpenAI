**3. Now, run your model on validation/test set and calculated the mean squared error. Report the error.**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# msk = np.random.rand(len(df)) < 0.8

# train = df[msk]
# test = df[~msk]

# print(len(test), len(train))
```

```python
y = df['SalePrice']
x = df[['OverallQual','GrLivArea','GarageCars','GarageArea','TotalBsmtSF']]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
help(model.fit)
```

```python
linr_model = model.fit(x_train, y_train)
linr_model
```

```python
r_sq = model.score(x_train, y_train)
r_sq
```

```python
print(f"intercept: {model.intercept_}")
print()
print(f"slope: {model.coef_}")
```

```python
y_pred = model.predict(x_train)
y_pred
```

```python
y_pred = model.intercept_ + model.coef_ * x_train
y_pred
```
