## ➡️ Coversion Rate Prediction

**Using `creative_image.csv` we would like to predict a creativeId's conversion rate given its image features. Create a feature matrix and a target vector. Each row in feature matrix corresponds to a sample in the dataset and each column is associated with a certain feature that is extracted from ad images.**

```python
df1 = creative_image_df.merge(click_train_df)
df1
```

```python
df2 = df1.groupby(['creativeId'])[['clicked', 'imageFeatures']].agg('mean')
df2
```

```python
df2 = df2.reset_index()
df2
```

```python
df3 = df2.merge(creative_image_df)
df3
```

```python
type(df3.imageFeatures[0])
```

```python
import ast
df3['imageFeaturesList'] = df3['imageFeatures'].apply(lambda x: ast.literal_eval(x))
df3
```

```python
type(df3.imageFeaturesList[0])
```

```python
df3['imageFeaturesList'].str.len().value_counts()
```

```python
features_col = list(str(i) for i in range(512))
```

```python
df3[features_col] = pd.DataFrame(df3.imageFeaturesList.values.tolist(), index= df3.index)
df3
```

**Use a linear regression model to predict conversion rate given image features for each creativeId. Evaluate your model's performance using 5-fold cross-validation. Does adding polynomial features boost the performance? Use mean-squared-error as your evaluation metric.**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
```

```python
Xcols= list(str(i) for i in range(512))
```

```python
X = df3[Xcols+['clicked']]
X
```

```python
y = df3['clicked']
y
```

```python
X_train, X_eval, y_train, y_eval = train_test_split(X, y, test_size=.2)
```

```python
len(y_train), len(y_eval)
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_train_predict = lr_model.predict(X_train)
```

```python
mse(y_true=y_train, y_pred=y_train_predict)
```

```python
y_eval_predict = lr_model.predict(X_eval)
mae(y_true=y_eval, y_pred=y_eval_predict)
```

```python
from sklearn.model_selection import cross_validate
```

```python
scores = cross_validate(LinearRegression(), X, y, scoring=('neg_mean_squared_error'), return_train_score=True)
```

```python
scores
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(2)
poly_features = poly.fit_transform(X_train)
model = LinearRegression().fit(poly_features, y_train)
mse(y_true= y_train, y_pred= model.predict(poly.fit_transform(X_train)))
```

```python
poly_features = poly.fit_transform(X_eval)
model = LinearRegression().fit(poly_features, y_eval)
mse(y_true= y_eval, y_pred= model.predict(poly.fit_transform(X_eval)))
```
